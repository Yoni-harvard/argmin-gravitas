```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("dark")
```

```python
actualPerformance = np.array([2, 8])
estimatedPerformance = np.array([5, 9])

fig, ax = plt.subplots(figsize=(7.5, 7.5))
plt.scatter(actualPerformance, estimatedPerformance, label="self-estimated")
plt.scatter(actualPerformance, actualPerformance, color="orange", label="actual")

plt.plot([2, 2], [2.11, 5], color="gray", linewidth=1)
plt.plot([8, 8], [8.1, 9], color="gray", linewidth=1)
ax.text(2.15, 3.4, "L", fontsize=20)
ax.text(8.15, 8.4, "U", fontsize=20)


#fig.suptitle("1. Original claim: $L > U$\n\n", fontsize=20)
plt.ylim(0, 10)
plt.xlim(0, 10)
plt.xlabel("Actual performance - $P$", fontsize=18)
plt.ylabel("Self-described performance - $\hat{P}$", fontsize=18)
plt.legend(loc="upper left")
plt.savefig("dk_claim_1.png")
plt.show()
```

```python
x = np.linspace(-12, 10, 1000)
y = x**2 + 2*x + 10


ax=plt.gca()
ax.set_xticks([])
ax.set_yticks([])
plt.xlabel("Actual performance = $P$", fontsize=14)
plt.ylabel("Self-described performance = $\hat{P}$", fontsize=14)
plt.plot(x, y)
```

```python
actualPerformance = np.array([2, 8])
estimatedPerformance = np.array([9, 8])

fig, ax = plt.subplots(figsize=(7.5, 7.5))
plt.scatter(actualPerformance, estimatedPerformance, label="self-estimated")
plt.scatter(actualPerformance, actualPerformance, color="orange", label="actual")

plt.plot([2, 2], [2.11, 9], color="gray", linewidth=1)
#plt.plot([8], [8], color="gray", linewidth=1)
ax.text(2.15, 2, "L", fontsize=16)
ax.text(8.15, 8, "H", fontsize=16)

fig.suptitle("$\hat{P}(L) > P(H)$\n\n", fontsize=20)
plt.ylim(0, 10)
plt.xlim(0, 10)
plt.xlabel("Actual performance - $P$", fontsize=18)
plt.ylabel("Self-described performance - $\hat{P}$", fontsize=18)
plt.legend(loc="lower right")
plt.savefig("dk_claim_4.png")
plt.show()
```
